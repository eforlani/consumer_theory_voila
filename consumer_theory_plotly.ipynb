# Consumer Theory Interactive Demo

Questo notebook mostra graficamente:
- Curve di indifferenza Cobb-Douglas a livelli fissi di utilità
- Vincolo di bilancio
- Paniere ottimale del consumatore

**Istruzioni:**
- Usa gli slider sotto per cambiare reddito (m), prezzi (px, py) e preferenze (alpha)
- Il grafico si aggiornerà automaticamente

**Funzione di utilità Cobb-Douglas:**
$$U(x,y) = x^\alpha \cdot y^{1-\alpha}$$

**Vincolo di bilancio:**
$$p_x \cdot x + p_y \cdot y = m$$

```python
# Configurazione per Voilà con Plotly
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import plotly.graph_objects as go
from ipywidgets import interact, FloatSlider

print("✓ Librerie caricate con successo")
```

```python
def utility(X, Y, alpha=0.5):
    """Cobb-Douglas utility function"""
    with np.errstate(divide='ignore', invalid='ignore'):
        return (X**alpha) * (Y**(1 - alpha))

def optimum_bundle(m=100, px=1, py=1, alpha=0.5):
    """Calculate optimal consumption bundle"""
    x_star = alpha * m / px
    y_star = (1 - alpha) * m / py
    return x_star, y_star

def plot_consumer(m=100, px=1, py=1, alpha=0.5):
    """
    Plot indifference curves, budget line, and optimal bundle using Plotly.
    """
    # Parametri fissi
    x_max = 160
    y_max = 160
    u_levels = [10, 20, 30, 40, 50, 60]
    
    # Create grid
    x = np.linspace(0.1, x_max, 150)
    y = np.linspace(0.1, y_max, 150)
    X, Y = np.meshgrid(x, y)

    # Calculate utility on grid
    U = utility(X, Y, alpha)

    # Budget line: y = (m - px*x)/py
    x_budget = np.linspace(0, m/px, 100)
    y_budget = (m - px * x_budget) / py

    # Optimal bundle
    x_star, y_star = optimum_bundle(m, px, py, alpha)
    u_star = utility(x_star, y_star, alpha)

    # Create Plotly figure
    fig = go.Figure()

    # Add indifference curves
    fig.add_trace(go.Contour(
        x=x,
        y=y,
        z=U,
        contours=dict(
            start=min(u_levels),
            end=max(u_levels),
            size=(max(u_levels)-min(u_levels))/(len(u_levels)-1),
            showlabels=True,
            labelfont=dict(size=10, color='white')
        ),
        colorscale='Blues',
        showscale=False,
        line=dict(width=2),
        name='Curve di Indifferenza'
    ))

    # Add budget line
    fig.add_trace(go.Scatter(
        x=x_budget,
        y=y_budget,
        mode='lines',
        line=dict(color='red', width=3),
        name='Vincolo di Bilancio',
        hovertemplate='x=%{x:.2f}<br>y=%{y:.2f}<extra></extra>'
    ))

    # Add optimal point
    if 0 <= x_star <= x_max and 0 <= y_star <= y_max:
        fig.add_trace(go.Scatter(
            x=[x_star],
            y=[y_star],
            mode='markers',
            marker=dict(size=15, color='red', symbol='circle',
                       line=dict(width=2, color='black')),
            name=f'Ottimo: ({x_star:.1f}, {y_star:.1f})',
            hovertemplate=f'<b>Paniere Ottimale</b><br>x*={x_star:.2f}<br>y*={y_star:.2f}<br>U*={u_star:.2f}<extra></extra>'
        ))

    # Layout
    fig.update_layout(
        title=dict(
            text='<b>Scelta del Consumatore: Curve di Indifferenza & Vincolo di Bilancio</b>',
            font=dict(size=16)
        ),
        xaxis=dict(
            title='<b>Bene X</b>',
            range=[0, x_max],
            showgrid=True,
            gridcolor='lightgray',
            zeroline=True
        ),
        yaxis=dict(
            title='<b>Bene Y</b>',
            range=[0, y_max],
            showgrid=True,
            gridcolor='lightgray',
            zeroline=True
        ),
        width=800,
        height=600,
        hovermode='closest',
        legend=dict(
            x=0.7,
            y=0.98,
            bgcolor='rgba(255,255,255,0.8)',
            bordercolor='black',
            borderwidth=1
        ),
        plot_bgcolor='white'
    )

    fig.show()

    # Print analytical results
    slope = -px / py
    x_int = m / px
    y_int = m / py
    
    print("="*65)
    print("📊 PARAMETRI")
    print("="*65)
    print(f"  💰 Reddito (m) = €{m:.2f}")
    print(f"  🏷️  Prezzo bene X (px) = €{px:.2f}")
    print(f"  🏷️  Prezzo bene Y (py) = €{py:.2f}")
    print(f"  ⚖️  Parametro preferenze (α) = {alpha:.2f}")
    print()
    print("📈 VINCOLO DI BILANCIO")
    print("="*65)
    print(f"  Equazione: {px:.2f}·x + {py:.2f}·y = {m:.2f}")
    print(f"  Forma esplicita: y = {y_int:.2f} - {px/py:.2f}·x")
    print(f"  Pendenza = {slope:.3f}")
    print(f"  Intercetta asse X = {x_int:.2f}")
    print(f"  Intercetta asse Y = {y_int:.2f}")
    print()
    print("✨ PANIERE OTTIMALE")
    print("="*65)
    print(f"  x* = {x_star:.2f} unità")
    print(f"  y* = {y_star:.2f} unità")
    print(f"  Utilità massima U* = {u_star:.2f}")
    print(f"  Spesa totale = €{px*x_star + py*y_star:.2f}")
    print(f"  Quota spesa in X: {(px*x_star/m)*100:.1f}%")
    print(f"  Quota spesa in Y: {(py*y_star/m)*100:.1f}%")
    print("="*65)

print("✓ Funzioni definite con successo")
```

## 🎛️ Grafico Interattivo

Usa gli slider qui sotto per esplorare come cambiano:
- 📉 La linea di bilancio (variando **m**, **px**, **py**)
- 🎯 Il paniere ottimale (punto rosso)
- 💭 Le preferenze del consumatore (variando **alpha**)

**Suggerimenti:**
- Aumenta il reddito (m) → la linea di bilancio si sposta verso l'esterno
- Aumenta px → la linea diventa più piatta (pivot sull'asse Y)
- Aumenta alpha → il consumatore preferisce più X rispetto a Y

**Interattività del grafico:**
- Passa il mouse sui punti per vedere i valori esatti
- Usa la toolbar in alto a destra per zoom, pan, reset

```python
interact(
    plot_consumer,
    m=FloatSlider(value=100, min=20, max=200, step=10, 
                  description="💰 Reddito (m):", continuous_update=False, 
                  style={'description_width': '120px'}),
    px=FloatSlider(value=1.0, min=0.5, max=5.0, step=0.1, 
                   description="🏷️ Prezzo X (px):", continuous_update=False,
                   style={'description_width': '120px'}),
    py=FloatSlider(value=1.0, min=0.5, max=5.0, step=0.1, 
                   description="🏷️ Prezzo Y (py):", continuous_update=False,
                   style={'description_width': '120px'}),
    alpha=FloatSlider(value=0.5, min=0.1, max=0.9, step=0.05, 
                      description="⚖️ Alpha (α):", continuous_update=False,
                      style={'description_width': '120px'})
);
```

# Consumer Preferences: Indifference Curves and the Budget Line

Goal: understand consumer choice as the interaction between:
- Preferences (indifference curves)
- Constraints (budget constraint)

We consider two goods:
- Good x (horizontal axis)
- Good y (vertical axis)

We use a simple Cobb-Douglas utility function:
\[
U(x,y) = x^\alpha y^{1-\alpha}
\]


```python
import numpy as np
import matplotlib.pyplot as plt

def utility(X, Y, alpha=0.5):
    return (X**alpha) * (Y**(1 - alpha))

def optimum_bundle(m=100, px=1, py=1, alpha=0.5):
    x_star = alpha * m / px
    y_star = (1 - alpha) * m / py
    return x_star, y_star

def plot_consumer(m=100, px=1, py=1, alpha=0.5,
                  x_max=160, y_max=160,
                  u_levels=(10, 20, 30, 40, 50, 60)):
    x = np.linspace(1e-3, x_max, 200)
    y = np.linspace(1e-3, y_max, 200)
    X, Y = np.meshgrid(x, y)

    y_budget = (m - px * x) / py
    x_star, y_star = optimum_bundle(m, px, py, alpha)
    u_star = utility(x_star, y_star, alpha)

    fig, ax = plt.subplots(figsize=(7, 5))

    ax.contour(X, Y, utility(X, Y, alpha), levels=list(u_levels))

    mask = (y_budget >= 0) & (y_budget <= y_max)
    ax.plot(x[mask], y_budget[mask], linewidth=2)

    if 0 <= x_star <= x_max and 0 <= y_star <= y_max:
        ax.scatter([x_star], [y_star], s=70)

    ax.set_xlim(0, x_max)
    ax.set_ylim(0, y_max)
    ax.set_xlabel("Good x")
    ax.set_ylabel("Good y")
    ax.set_title("Indifference Curves + Budget Line")
    plt.show()

    slope = -px / py
    x_int = m / px
    y_int = m / py
    print(f"m={m}, px={px}, py={py}, alpha={alpha}")
    print(f"Budget line: y = (m/py) - (px/py)x  | slope = {slope:.2f}")
    print(f"Intercepts: x-intercept m/px = {x_int:.2f}, y-intercept m/py = {y_int:.2f}")
    print(f"Optimal bundle: x*={x_star:.2f}, y*={y_star:.2f}, U*={u_star:.2f}")
```

```python
import ipywidgets as widgets
from IPython.display import display

m = widgets.FloatSlider(value=100, min=20, max=200, step=10, description="Income m", continuous_update=False)
px = widgets.FloatSlider(value=1.0, min=0.5, max=5.0, step=0.1, description="Price px", continuous_update=False)
py = widgets.FloatSlider(value=1.0, min=0.5, max=5.0, step=0.1, description="Price py", continuous_update=False)
alpha = widgets.FloatSlider(value=0.5, min=0.1, max=0.9, step=0.05, description="alpha", continuous_update=False)
x_max = widgets.FloatSlider(value=160, min=60, max=300, step=10, description="x window", continuous_update=False)
y_max = widgets.FloatSlider(value=160, min=60, max=300, step=10, description="y window", continuous_update=False)

out = widgets.Output()

def redraw(change=None):
    with out:
        out.clear_output(wait=True)
        plot_consumer(
            m=m.value,
            px=px.value,
            py=py.value,
            alpha=alpha.value,
            x_max=x_max.value,
            y_max=y_max.value,
            u_levels=(10, 20, 30, 40, 50, 60),
        )

for w in [m, px, py, alpha, x_max, y_max]:
    w.observe(redraw, names="value")

controls = widgets.VBox([m, px, py, alpha, x_max, y_max])
ui = widgets.HBox([controls, out])

display(ui)
redraw()
```
